Se crea un programa para reconocimiento de autos por medio de un video o imagenes, desarrollado en lenguaje de python. La cual se basa en entrenar el programa para que aprenda el reconocimiento del autos.

El entrenamiento se hizo mediante la herramienta de OpenCV(Opencv_createsamples), el cual crea un archivo vector.

Opencv_traincascade el entrenamiento, usando el vector creado por el opencv_createsamples y el conjunto de im√°genes negativas, dando como resultado, un archivo .xml de entrenamiento.

Este proceso se repite varias veces y durante horas para que mejore su reconocimiento y sea mas exacto en el momento de reconocer las imagenes que se le entregan al programa o imagenes.


```python
# librerias para el funcionamiento de este programa.
import cv2
import numpy as np
 
# Clasificador de entrenamiento para carros.
car_cascade = cv2.CascadeClassifier('cascade.xml')

# Captura de v√≠deo.
cap = cv2.VideoCapture(0)
 
# si el v√≠deo abrio, inicia el while.
while True:
    # Lee los cuadros por segundo del v√≠deo.
    ret, frames = cap.read()
     
    # Convierte a escala de grises, internamente.
    gray = cv2.cvtColor(frames, cv2.COLOR_BGR2GRAY)
     
 
    # Detecta diferentes tama√±os de im√°genes para asociarlos a un carro.
    cars = car_cascade.detectMultiScale(gray, 1.1, 1)
     
    # Crea un rectangulo donde encuentra un carro.
    for (x,y,w,h) in cars:
        cv2.rectangle(frames,(x,y),(x+w,y+h),(0,0,255),2)
 
   # Muestra en v√≠deo los cuadros por segundo (frame)
    print(cv2.imshow('video', frames))
    print(cv2.getBuildInformation())

    # Tecla ESC para cerrar.
    if cv2.waitKey(33) == 27:
        break
 
# Destruye todas las ventanas generadas por el programa-
cv2.destroyAllWindows()
```
